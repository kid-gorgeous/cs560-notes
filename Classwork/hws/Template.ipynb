```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# evenly sampled time at 200ms intervals
t = np.arange(1., 50,5)

#red dashes, blue squares and green triangles

#plt.axis([0,50,0,100])
plt.plot(t, t * np.log2(t) + 100, 'b', t, t*t, 'r^')
plt.show()
```
