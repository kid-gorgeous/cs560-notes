Pass the array to a merge sort function (not the buggy one) and record the total number of iterations the for loop executes.

````python
def merge_sort(arr):
    count = 0

    if len(arr) > 1:
        mid = len(arr)//2
        L = arr[:mid]
        R = arr[mid:]

        count += merge_sort(L)
        count += merge_sort(R)

        i = j = k = 0

        while i < len(L) and j < len(R):
            count += 1
            if L[i] < R[j]:
                arr[k] = L[i]
                i += 1
            else:
                arr[k] = R[j]
                j += 1
            k += 1

        while i < len(L):
            count += 1
            arr[k] = L[i]
            i += 1
            k += 1

        while j < len(R):
            count += 1
            arr[k] = R[j]
            j += 1
            k += 1

    return count

arr = [x for x in range(1<<10,0,-1)]
arr1 = [x for x in range(1<<12,0,-1)]
count = merge_sort(arr)
count1 = merge_sort(arr1)
print("Total number of iterations:", count)
print("Total number of interations: ",count1)
````
